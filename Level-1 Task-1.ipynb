Upload the dataset when prompted below. This notebook will:
- Show data dimensions
- Handle missing values
- Analyze target variable 'Aggregate rating'
- Check class imbalance

```python
# STEP 1: Upload CSV File
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))
df.head()
```

```python
# STEP 2: Dataset Dimensions
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
# STEP 3: Missing Values
missing = df.isnull().sum()
missing[missing > 0]
```

```python
# STEP 4: Handle Missing Values
df['Cuisines'] = df['Cuisines'].fillna("Not specified")
df = df.dropna(subset=['Longitude', 'Latitude'])
```

```python
# STEP 5: Convert Data Types
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
# STEP 6: Visualize Distribution of 'Aggregate rating'
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# STEP 7: Check Class Imbalance
sns.countplot(data=df, x='Aggregate rating')
plt.title("Aggregate Rating Class Distribution")
plt.show()
```

```python
# STEP 8: Save and Download Cleaned CSV
df.to_csv("Cleaned_Dataset.csv", index=False)
files.download("Cleaned_Dataset.csv")
```
