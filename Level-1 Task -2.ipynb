## Task: Descriptive Analysis

This notebook performs statistical and categorical analysis on the restaurant dataset.

```python
# STEP 1: Upload CSV File
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))
df.head()
```

### STEP 2: Descriptive Statistics for Numerical Columns

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Mean, Median, Std of specific numeric columns
print("Mean of Aggregate Rating:", df['Aggregate rating'].mean())
print("Median of Votes:", df['Votes'].median())
print("Std Dev of Cost for Two:", df['Average Cost for two'].std())
```

### STEP 3: Explore Categorical Variables

```python
# Country Code value counts
df['Country Code'].value_counts()
```

```python
# Top 10 cities
df['City'].value_counts().head(10)
```

```python
# Top 10 cuisines
df['Cuisines'].value_counts().head(10)
```

### STEP 4: Visualize Top Cities and Cuisines

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top cities
top_cities = df['City'].value_counts().head(10)
top_cities.plot(kind='bar', figsize=(10,5), title='Top 10 Cities with Most Restaurants')
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# Top cuisines
top_cuisines = df['Cuisines'].value_counts().head(10)
top_cuisines.plot(kind='bar', figsize=(10,5), color='orange', title='Top 10 Cuisines')
plt.ylabel("Number of Restaurants")
plt.show()
```
